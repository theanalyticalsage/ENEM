O Programa Universidade para Todos - Prouni tem como finalidade a concessão de bolsas de estudo integrais e parciais em cursos de graduação e sequenciais de formação específica, em instituições de ensino superior privadas. Criado pelo Governo Federal em 2004 e institucionalizado pela Lei nº 11.096, em 13 de janeiro de 2005 oferece, em contrapartida, isenção de tributos àquelas instituições que aderem ao Programa.

Dirigido aos estudantes egressos do ensino médio da rede pública ou da rede particular na condição de bolsistas integrais, com renda familiar per capita máxima de três salários mínimos, o Prouni conta com um sistema de seleção informatizado e impessoal, que confere transparência e segurança ao processo. Os candidatos são selecionados pelas notas obtidas no Exame Nacional do Ensino Médio - Enem conjugando-se, desse modo, inclusão à qualidade e mérito dos estudantes com melhores desempenhos acadêmicos.

O Programa possui também ações conjuntas de incentivo à permanência dos estudantes nas instituições, como a Bolsa Permanência e ainda o Fundo de Financiamento Estudantil - Fies, que possibilita ao bolsista parcial financiar parte da mensalidade não coberta pela bolsa do programa.

O Prouni já atendeu, desde sua criação até o processo seletivo do segundo semestre de 2018, mais de 2,47 milhões de estudantes, sendo 69% com bolsas integrais.

O Programa Universidade para Todos, somado ao Fies, ao Sistema de Seleção Unificada (Sisu), ao Programa de Apoio a Planos de Reestruturação e Expansão das Universidades Federais (Reuni), a Universidade Aberta do Brasil (UAB) e a expansão da rede federal de educação profissional e tecnológica ampliam significativamente o número de vagas na educação superior, contribuindo para um maior acesso dos jovens à educação superior.

# Desenharei o perfil de 3 cursos

**Devido ao resultado da Análise exploratória de dados escolherei os cursos Pedagogia, Administração e Direito.**

Estão com os dados Completos

Para traçar o perfil, utilizarei uma técnica de escala multidimensional, atribuirei uma 'distancia' de similaridades para cada registro e reduzirei todas as dimensões a duas somente, eixo horizontal e eixo vertical, dado este cenário utilizarei um método de clusterização hierarquico para agrupar estas distancias por similaridade, no final explicarei os perfis.

Separarei os tres cursos escolhidos conforme abaixo:

```python
df_adm=dados_clean[dados_clean['nome']=="AdministraC'C#o"]
df_ped=dados_clean[dados_clean['nome']=="Pedagogia"]
df_dir=dados_clean[dados_clean['nome']=="Direito"]
```

Selecionarei a seguir as variáveis 'Features' que levarei em consideração para calcular as similaridades

```python
features=['mensalidade' , 'bolsa_integral_cotas', 'bolsa_integral_ampla', 'bolsa_parcial_ampla'
          , 'nota_integral_ampla', 'nota_integral_cotas', 'nota_parcial_ampla','nota_parcial_cotas']
```

importarei bibliotecas para auxiliar no processo

```python
from sklearn.preprocessing import StandardScaler
from sklearn import manifold
from sklearn.metrics import euclidean_distances
from sklearn.manifold import MDS
```

devido a grande diferença entre as escalas de valores, notas, mensalidades e bolsas, normalizarei para auxiliar

```python
scaler = StandardScaler()
sc_df_adm=scaler.fit_transform(df_adm[features])
sc_df_ped=scaler.fit_transform(df_ped[features])
sc_df_dir=scaler.fit_transform(df_dir[features])

df_adm_sc=pd.DataFrame(sc_df_adm, columns=[features])
df_ped_sc=pd.DataFrame(sc_df_ped, columns=[features])
df_dir_sc=pd.DataFrame(sc_df_dir, columns=[features])
```

Criando o modelo de MDS

```python
df_adm_sc.shape[1]
```

```python
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist, jaccard
from scipy.spatial import distance
```

```python
res=pdist(df_adm_sc, 'jaccard')
```

```python
df_adm_sc
```

```python
distance = pd.DataFrame(squareform(res), index=df_adm_sc.index, columns= df_adm_sc.index)
```

```python
distance
```

```python
res.shape[0]/164*8
```

```python
embedding = MDS(n_components=2)
```

```python
adm= embedding.fit_transform(sc_df_adm)
ped= embedding.fit_transform(sc_df_ped)
dir= embedding.fit_transform(sc_df_dir)
```

Calculando as dissimilaridades

```python
dissimilaridades_adm=pd.DataFrame(euclidean_distances(adm))
dissimilaridades_ped=pd.DataFrame(euclidean_distances(ped))
dissimilaridades_dir=pd.DataFrame(euclidean_distances(dir))
```

```python
seed=1234
```

Atribuindo o modelo

```python
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=seed,
                   dissimilarity="precomputed", n_jobs=1)
```

```python
pos_adm = mds.fit(dissimilaridades_adm).embedding_
pos_ped = mds.fit(dissimilaridades_ped).embedding_
pos_dir = mds.fit(dissimilaridades_dir).embedding_
```

### Resultado da Escala Multidimensional

```python
fig = plt.figure(figsize=(15,15))

#Administração
ax3 = fig.add_subplot(331)
ax3.set_xlabel('Dimensão Horizontal')
ax3.set_ylabel('Dimensão Vertical')
ax3.set_title('Disp. do Curso de Adm')
plt.scatter(pos_adm[:, 0], pos_adm[:, 1], color='green', lw=0, label='MDS')

#Pedagogia
ax4 = fig.add_subplot(332)
ax4.set_xlabel('Dimensão Horizontal')
ax4.set_title('Disp. do Curso de Pedagogia')
plt.scatter(pos_ped[:, 0], pos_ped[:, 1], color='blue', lw=0, label='MDS')


#Direito
ax4 = fig.add_subplot(333)
#ax4.axes([0., 0., 1., 1.])
ax4.set_xlabel('Dimensão Horizontal')
#ax4.set_ylabel('Dimensão Vertical')
ax4.set_title('Disp. do Curso de Direito')
plt.scatter(pos_dir[:, 0], pos_dir[:, 1], color='red', lw=0, label='MDS')
```

```python
df_adm['Eixo_1']=pos_adm[:,0]
df_adm['Eixo_2']=pos_adm[:,1]
df_ped['Eixo_1']=pos_ped[:,0]
df_ped['Eixo_2']=pos_ped[:,1]
df_dir['Eixo_1']=pos_dir[:,0]
df_dir['Eixo_2']=pos_dir[:,1]
```

Criando a Clusterização hierarquica

```python
from sklearn.cluster import AgglomerativeClustering

cluster_adm = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster_ped = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster_dir = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
c_adm=cluster_adm.fit_predict(pos_adm)
c_ped=cluster_ped.fit_predict(pos_ped)
c_dir=cluster_dir.fit_predict(pos_dir)
```

```python
df_adm['Cluster']=c_adm   
df_ped['Cluster']=c_ped
df_dir['Cluster']=c_dir
```

Mostrando o Resultado da Clusterização

```python
fig = plt.figure(figsize=(15,15))

#Administração
ax3 = fig.add_subplot(331)
ax3.set_xlabel('Dimensão Horizontal')
ax3.set_ylabel('Dimensão Vertical')
ax3.set_title('Disp. do Curso de Adm')
scatter_adm=plt.scatter(df_adm['Eixo_1'], df_adm['Eixo_2'], c=df_adm['Cluster'],cmap='rainbow',  lw=0, label='MDS')
legend0 = ax3.legend(*scatter_adm.legend_elements(),
                    loc="upper left", title='Cluster')
ax3.add_artist(legend0)

#Pedagogia
ax4 = fig.add_subplot(332)
ax4.set_xlabel('Dimensão Horizontal')
ax4.set_title('Disp. do Curso de Pedagogia')
scatter_ped=plt.scatter(df_ped['Eixo_1'], df_ped['Eixo_2'], c=df_ped['Cluster'],cmap='rainbow',  lw=0, label='MDS')
legend1 = ax4.legend(*scatter_ped.legend_elements(),
                    loc="upper left", title='Cluster')
ax4.add_artist(legend1)

#Direito
ax4 = fig.add_subplot(333)
ax4.set_xlabel('Dimensão Horizontal')
ax4.set_title('Disp. do Curso de Direito')
scatter_dir=plt.scatter(df_dir['Eixo_1'], df_dir['Eixo_2'], c=df_dir['Cluster'],cmap='rainbow',  lw=0, label='MDS')
legend2 = ax4.legend(*scatter_dir.legend_elements(),
                    loc="upper left", title='Cluster')
ax4.add_artist(legend2)

```
